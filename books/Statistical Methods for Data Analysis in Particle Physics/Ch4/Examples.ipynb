```python
import numpy as np
import mplhep
import matplotlib.pyplot as plt

mplhep.style.use("ATLAS")
```

# Example 4.14

```python
r = np.random.random(size=100000)
x = -np.log(1 - r)
```

```python
bin_edges = np.linspace(0, 5, 51)
fig, ax = plt.subplots()
ax.hist(x, bins=bin_edges, density=True, label="dist. random numbers")
ax.plot(bin_edges, np.exp(-bin_edges), label="theory formular")
```
